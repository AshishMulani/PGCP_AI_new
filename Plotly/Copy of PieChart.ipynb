```python
import plotly.express as px
```

```python
df = px.data.tips()
df
```

```python
fig=px.pie(df, names = 'day', values='total_bill')
fig
```

```python
fig=px.pie(df, names = 'day', values='total_bill', hole=0.5,template='seaborn')
fig
```

```python
fig=px.sunburst(df, path =['day', 'sex','time'], values='total_bill', template='seaborn')
fig.update_traces(texttemplate='%{label}<br>%{percentRoot:.1%}')
fig
```

```python
fig=px.treemap(df, path =['day', 'sex','time'], values='total_bill', template='seaborn')
fig.update_traces(texttemplate='%{label}<br>%{percentRoot:.1%}')
fig
```
