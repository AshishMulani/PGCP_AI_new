```python
import plotly.express as px
import seaborn as sns
```

```python
df = sns.load_dataset('diamonds')
df
```

```python
df_s = df.sample(100)
df_s
```

```python
fig=px.scatter(df_s, x='carat', y = 'price')
fig
```

```python
fig=px.scatter(df_s, x='carat', y = 'price', color='cut', symbol='clarity')
fig
```

```python
fig=px.scatter(df_s, x='carat', y ='price', color='cut', size='carat')
fig
```
