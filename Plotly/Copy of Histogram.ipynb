```python
import plotly.express as px
```

```python
df = px.data.iris()
df
```

```python
fig=px.histogram(df , x='sepal_length')
fig
```

```python
fig=px.histogram(df , x='sepal_length', color='species', barmode='overlay')
fig
```

```python
fig=px.histogram(df , x='sepal_length', color='species', barmode='overlay', facet_row='species')
fig
```

```python
fig=px.histogram(df , x='sepal_length', color='species', barmode='overlay', facet_col='species')
fig
```
