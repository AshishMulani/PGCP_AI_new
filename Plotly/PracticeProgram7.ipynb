```python
'''Topic : Plotly

1.	Draw a bar chart showing GDP percentage for European countires. [use gdpminder dataset]
	Use different semantics as orientation, bar mode, bar norm, text to improve your chart
	For the same dataset show GDP over the years using animation frame as Country '''
```

```python
import plotly.express as px
```

```python
df = px.data.gapminder()
df.head()
```

```python
filt = df['continent']=='Europe'
df_f = df.loc[filt]
df_f.head()
```

```python
fig=px.bar(df_f, x='country', y='gdpPercap',orientation='v',text='gdpPercap',color='country')
fig
```

```python
fig=px.bar(df_f, x='country', y='gdpPercap', color='country',orientation='v',text='gdpPercap',animation_frame='year')
fig.update_layout(title='GDP Percentage per country',
                 showlegend = True,
                 hovermode='x unified',                        
                barmode='stack',     
                barnorm='percent'
                 )
#Use different semantics as orientation, bar mode, bar norm, text to improve your chart
```

```python
'''2.	Draw a Histogram showing votes secured by 3 candidates - Coderre, Joly, Bergeron
	Use semantics -bar mode, color, facet_row to improve your chart. Create a cumulative graph as well. [use election dataset]'''
```

```python
df = px.data.election()
df.head()
```

```python
fig=px.histogram(df , x=['Coderre','Bergeron','Joly'],color='winner', barmode='overlay')
fig
```

```python
fig=px.histogram(df , x=['Coderre','Bergeron','Joly'],color='winner', barmode='overlay',facet_row='winner')
fig
```
