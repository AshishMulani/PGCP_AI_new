```python
import plotly.express as px
```

```python
df = px.data.gapminder()
df
```

```python
filt = df['continent']=='Asia'
df_f = df.loc[filt]
df_f
```

```python
fig=px.bar(df_f, x='country', y='lifeExp')
fig
```

```python
fig=px.bar(df_f, x='country', y='lifeExp', color='country', animation_frame='year')
fig.update_layout(title='Life Expentency per country- year 2007',
                 showlegend = True,
                 hovermode='y unified')
```

```python
filt = (df['continent']=='Asia') & (df['year']==2007)
df_f = df.loc[filt]
df_f
```

```python
fig=px.bar(df_f, x='country', y='lifeExp')
fig
```

```python
fig=px.bar(df_f, x='country', y='lifeExp', color='country')
fig.update_layout(title='Life Expentency per country- year 2007',
                 showlegend = True,
                 hovermode='y unified')
```

```python
fig=px.bar(df, x='country', y='lifeExp', color='country', animation_frame='year',
          facet_row='continent')
fig.update_layout(title='Life Expentency per country- year 2007',
                 showlegend = True,
                 hovermode='y unified')
```
