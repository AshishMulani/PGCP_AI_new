```python
import plotly.express as px
```

```python
dir(px.data)
```

```python
df = px.data.stocks()
df
```

```python
fig=px.line(df, x='date', y=['GOOG','AAPL','AMZN'])
fig
```

```python
fig=px.line(df, x='date', y=['GOOG','AAPL','AMZN'], markers=True)
fig.update_layout(xaxis_title='Month',
                 yaxis_title='Stock Price',
                 title = 'Stock Prices for Google. Apple & Amazon',
                 showlegend=True,
                 plot_bgcolor='#DBE68C',
                 hovermode='x unified')
```
