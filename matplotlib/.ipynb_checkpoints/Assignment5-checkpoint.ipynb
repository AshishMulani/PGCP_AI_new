```python
'''Topic: Matplotlib

1.	Draw line chart to show urban and rural population of India over the years using population.csv 
    [rural 	population is the difference between total population and urban population]
Note: use appropriate labels, title and legend for each chart'''
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv('population.csv')
df
```

```python
df['Rural Population'] = df['Population'] - df['Urban Population']
df
```

```python
df.dropna()
```

```python
plt.plot(df['Year'], df['Rural Population'], color = "#F54927", lw=0.8, ls='--', label='Rural Population of India',marker='o')
plt.plot(df['Year'], df['Urban Population'], color = "#27F5E0", lw=0.8, ls='-.', label='Urban Population of India',marker='s')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.title('Rural Population Vs Urban Population over the years')
plt.show()
```

```python
#2.Using a Bar chart display India's total and rural population from year 2020 to 2025
```

```python
filt_df = df[(df['Year'] >= 2020) & (df['Year'] <= 2025)]
filt_df
```

```python
x_vals = np.arange(len(filt_df['Year']))
width=0.4
x_vals
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
plt.bar(x_vals - width/2, filt_df['Population'],width=width, color = '#a86232', label='Total Population')
plt.bar(x_vals + width/2, filt_df['Rural Population'],width=width, color ='#324ca8', label='Rural Population')

plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(x_vals, filt_df['Year'])
plt.legend()
plt.title('Rural Population Vs Total Population')
plt.show()
```

```python
#2.	Using gapminder2007.csv show life expectancy of top 10 highly populated countries from Asia
```

```python
df=pd.read_csv('gapminder2007.csv')
df
```

```python
new_df = df[df['continent'] == 'Asia'].sort_values('pop',ascending=False).head(10)
new_df
```

```python
plt.bar(new_df['country'], new_df['lifeExp'], width = 0.4, label ='Life Expectancy')
plt.xlabel('Country')
plt.xticks(rotation =90)
plt.legend()
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy of Top 10 Most Populated Countries')
plt.legend()
plt.show()
```

```python
#3.	Using a pie chart show distribution of top six selling cars [Create data as required]
```

```python
data = {'Car Model': ['Swift', 'Baleno', 'Nexon', 'WagonR', 'Creta', 'Punch'],
        'Sales': [18500, 17200, 15800, 15500, 14900, 14200]}
df = pd.DataFrame(data)
df
```

```python
plt.pie(df['Sales'], 
        labels=df['Car Model'],
        colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6'],
        shadow=True,
        startangle=140, 
        explode=[0, 0, 0, 0, 0.1,0],
        wedgeprops={'edgecolor':'Black'},
        autopct='%1.1f%%')

plt.title('Distribution Selling Car Models', fontsize=15, fontweight='bold')
plt.legend(df['Car Model'], title="Models", loc="center left", bbox_to_anchor=(1.5, 0, 0.5, 1))
plt.show()
```

```python
#4.	Using Histogram show distribution of developers by their experience [use survey_data_sample.csv]
```

```python
df = pd.read_csv('survey_results-sample.csv')
df
```

```python
df=df.dropna()
df
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
plt.hist(df['YearsCode'], bins=50, color='#69b3a2', edgecolor='black',label ="Developers")

plt.title('Distribution of Developers by Experience (YearsCode)', fontweight='bold')
plt.xlabel('Years of Coding Experience')
plt.ylabel('Number of Developers')
plt.xticks(rotation =90)
plt.legend()

plt.show()
```
