```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
year = [1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020, 2025]
p_india = [772647793, 864972221, 960301044, 1057922733, 1154676322, 1243481564, 1328024498, 1402617695, 1463865525]
p_urban = [190321782, 222296728, 255558824, 291350282, 334479406, 380744554, 429069459, 483098640, 542742539]
```

```python
plt.plot(year, p_india)
plt.show()
```

```python
plt.plot(year, p_india, color = "#F54927", lw=0.8, ls='--', label='Population of India')
plt.xlabel('Year')
plt.ylabel('Population of India')
plt.legend()
plt.title('Population of India over the years')
plt.show()
```

```python
plt.plot(year, p_india, color = "#F54927", lw=0.8, ls='--', label='Population of India')
plt.plot(year, p_urban, color = "#27F5E0", lw=0.8, ls='-.', label='Urban Population of India')
plt.xlabel('Year')
plt.ylabel('Population of India')
plt.legend()
plt.title('Population of India Vs Urban Population over the years')
plt.show()
```
