```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
count = [9354, 8837, 7456, 5806, 5638]
langs = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
```

```python
plt.pie(count, labels = langs)
plt.show()
```

```python
plt.pie(count, labels = langs,
       colors =['Red','Blue','Pink','Green','Yellow'],
       explode=[0, 0, 0, 0.1, 0],
       shadow=True,
       startangle=90,
       wedgeprops={'edgecolor':'Black'},
       autopct="%1.1f%%")
plt.show()
```
