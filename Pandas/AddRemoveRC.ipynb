```python
import pandas as pd
```

```python
data = {'FullName' :['Rucha N','Rohit K', 'Kedar P', 'Ankita G'],
       'Email' : ['Rucha07@gmail.com', 'Rohit.10@gmail.com', 'KedarTennis@gmail.com','Ankita21@gmail.com'],
        'Salary':[100000, 120000,90000,75000]
       }
df = pd.DataFrame(data)
df
```

```python
df['Bonus'] = df['Salary']*0.1
df
```

```python
df[['FirstName', 'LastName']] = df['FullName'].str.split(' ', expand=True)
df
```

```python
df['Age'] = [23,24,20,25]
df
```

```python
df.drop(columns=['LastName'],inplace=True)
df
```

```python
row_data = {'FullName':'Parth P','Email': 'Parth11@gmail.com',
           'Salary':50000, 'Bonus': 5000.0, 
           'FirstName': 'Parth', 'Age':33}
df=pd.concat([df, pd.DataFrame([row_data])], ignore_index=True)
df
```

```python
row_data = {'FullName':['Parth P', 'Priya M'],'Email': ['Parth11@gmail.com', 'Priya01@gmail.com'],
           'Salary':[50000, 60000], 'Bonus': [5000.0, 6000.0], 
           'FirstName': ['Parth', 'Priya'], 'Age':[33,30]}
df1 = pd.DataFrame(row_data)
df=pd.concat([df,df1], ignore_index=True)
df
```

```python
df.drop(index=[3], inplace=True)
df
```

```python
df.drop_duplicates(inplace=True)
df
```
