```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
sns.kdeplot(df,x='total_bill')
```

```python
sns.kdeplot(df,x='total_bill', bw_adjust=0.25)
```

```python
sns.kdeplot(df,x='total_bill', hue='time')
```

```python
sns.kdeplot(df,x='total_bill', hue='time', multiple='stack')
```
