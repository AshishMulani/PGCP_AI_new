```python
import seaborn as sns
```

```python
'''Topic : Seaborn

1.	Draw a histogram showing distribution of flipper_length across species for penguin dataset
	Create a stack plot and step plot- – Write down your interpretations'''
```

```python
df= sns.load_dataset("penguins")
df
```

```python
df=df.dropna()
```

```python
sns.histplot(data=df, x="flipper_length_mm", hue="species", bins=20)
```

Interpretations:
    1)The histogram shows distribution  of flipper length among different species. 
    2) Gentoo has the largest flipper lengths and Adelie has smallest whereas Chinstrap has the inbetween lengths.

```python
sns.histplot(data=df, x="flipper_length_mm", hue="species", bins=20,element='step')
```

Step plot shows the distribution without filled bars making comparison clearer.
It makes easier to compare overlaping distribution.

```python
sns.histplot(data=df, x="flipper_length_mm", hue="species", bins=20,multiple='stack')
```

It shows combined distribution of flipper lengths.

```python

'''2.	Draw a bivariant KDE plot to show relationship between horsepower and displacement 
[use mpg dataset] – Write down your interpretations'''
```

```python
df = sns.load_dataset('mpg')
df
```

```python
df=df.dropna()
```

```python
sns.kdeplot(df,x='horsepower', y='displacement', fill=True)
```

1)The plot shows positive linear relationship between displacement and horsepower.
2)As displacement increases, horsepower also increases.
3)The dark region indicates most cars falls in that region where displacement and horsepower is low.
4)There is very few cars having large displacement and horsepower.
