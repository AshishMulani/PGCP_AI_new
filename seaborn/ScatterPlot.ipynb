```python
import seaborn as sns
```

```python
df = sns.load_dataset('diamonds')
df
```

```python
sns.scatterplot(df.sample(100), x='carat', y='price')
```

```python
sns.scatterplot(df.sample(100), x='carat', y='price', hue='cut')
```

```python
sns.scatterplot(df.sample(100), x='carat', y='price', hue='cut', size='clarity')
```

```python
sns.scatterplot(df.sample(100), x='carat', y='price', hue='cut', style='cut')
```

```python
sns.scatterplot(df.sample(100), x='carat', y='price', hue='cut', style='cut', alpha=0.5)
```
