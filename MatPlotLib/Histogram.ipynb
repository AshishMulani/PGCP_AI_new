```python
df = pd.read_csv('survey_results-sample.csv')
df
```

```python
df.dropna()
```

```python
df[['Min Age', 'Max Age']] = df['Age'].str.extract(r'(\d+)-(\d+)').astype(float)
```

```python
df
```

```python
df.dropna()
```

```python
df['Mean Age'] = df[['Min Age', 'Max Age']].mean(axis=1)
df=df.dropna()
df
```

```python
bins = [20, 30, 40, 50, 60]
plt.hist(df['Mean Age'], bins=bins, edgecolor='Black', label ="Respondent's Age")
plt.xlabel("Respondent's Age")
plt.ylabel("Count")
plt.legend()
plt.title("Survey Respondents by Age")
plt.show()
```
